# Example literature analysis

This notebook demonstrates the end-to-end workflow:

1. **Search** PubMed for a topic
2. **Fetch** article summaries
3. **Analyze** publications per year and top journals
4. **Visualize** the results

The search/fetch steps need network access (and ideally an `NCBI_API_KEY` in `.env`).
If you're offline, skip to the *Analysis on sample data* section, which uses a
hard-coded list and runs anywhere.

```python
from research.analysis import plot_publications_per_year, summarize, top_journals
from research.literature import fetch_summaries, search_pubmed
```

## 1. Search and fetch (requires network)

```python
query = "CRISPR off-target effects"
pmids = search_pubmed(query, retmax=50)
articles = fetch_summaries(pmids)
print(f"Fetched {len(articles)} articles for {query!r}")
```

## 2. Summarize

```python
summary = summarize(articles)
summary
```

```python
top_journals(articles, n=10)
```

## 3. Visualize

```python
fig = plot_publications_per_year(articles)
fig
```

## Analysis on sample data (offline)

Runs without network access, using a small hand-built list of `Article` objects.

```python
from research.literature import Article

sample = [
    Article(pmid="1", title="Genome editing advances", journal="Nature", year="2021"),
    Article(pmid="2", title="Off-target detection", journal="Nature", year="2023"),
    Article(pmid="3", title="Base editing", journal="Cell", year="2023"),
    Article(pmid="4", title="Prime editing", journal="Science", year="2024"),
    Article(pmid="5", title="Delivery methods", journal="Cell", year="2024"),
]

print(summarize(sample))
plot_publications_per_year(sample)
```
